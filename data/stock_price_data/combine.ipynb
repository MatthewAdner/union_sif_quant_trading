```python
def categorize_market_cap(value):
    if value >= 200e9:
        return 'mega'
    elif 10e9 <= value < 200e9:
        return 'large'
    elif 2e9 <= value < 10e9:
        return 'mid'
    elif 250e6 <= value < 2e9:
        return 'small'
    elif 0 < value < 250e6:
        return 'micro'
    # if there isn't data
    else:
        return np.nan
```

```python
size_category_slice = combined_from_parq.xs('Market_Cap',level=0, axis=1).map(categorize_market_cap)
size_category_slice.columns = pd.MultiIndex.from_product([['Size_Category'], size_category_slice.columns])

display(size_category_slice)

combined_from_parq = pd.concat([combined_from_parq, size_category_slice], axis=1)
display(combined_from_parq)
```
